### Importing all libraries required

```python
# Data Manipulation Libraries
import numpy as np
import pandas as pd

# Visualisation Libraries 
import matplotlib.pyplot as plt
import seaborn as sns 

# Error Handling Library
import warnings 
warnings.filterwarnings("ignore")
```

### Reading the dataset

```python
# Reading CSV data
df = pd.read_csv("SampleSuperstore.csv")
df
```

### Exploring dataset

```python
df.shape
```

The dataset consists of 13 features and each feature consists of 9994 samples.

```python
#getting the information abut dataset
df.info()
```

- There are no null values present in dataset.
- 'Postal Code','Sales','Quantity','Discount','Profit' are numerical features rest all are Categorical.

```python
#duplicated values
df.duplicated().sum()
```

```python
#dropping the duplicated rows
df.drop_duplicates()
```

```python
#Here postal code is not a useful column, hence we are dropping it.
#Since all the data from the United States we can drop also the Country column
df.drop([ 'Country','Postal Code'],axis=1, inplace=True)
df
```

```python
#correlation
sns.heatmap(df.corr() , annot=True)
```

It can be observe that the numeric variables are not correlated with each other since all the correlation coefficients are close to zero, however

* Discount and Sales are negatively correlated 
* Sales and Profit are neutrally correlated

We can analyse the data further in **3 different ways:**

### Sales Vs Profit Vs Discount

                                      Based On Ship Mode

```python
sns.set(style="whitegrid", color_codes=True)
fig , ax = plt.subplots(1,3,figsize=(20,8))
sns.barplot(x='Ship Mode' , y= 'Sales', data=df , ax=ax[0] )
sns.barplot(x='Ship Mode' , y= 'Profit', data=df , ax=ax[1])
sns.barplot(x='Ship Mode' , y= 'Discount', data=df , ax=ax[2])
```

First Class has maximum profits due to high discount

                                       Based on Segment

```python
fig , ax = plt.subplots(1,3,figsize=(20,8))
sns.barplot(x='Segment' , y= 'Sales', data=df , ax=ax[0] )
sns.barplot(x='Segment' , y= 'Profit', data=df , ax=ax[1])
sns.barplot(x='Segment' , y= 'Discount', data=df , ax=ax[2])
```

even with the low Discounts the Home Office has the higher Sales and Profits

                                        Based on Region 

```python
fig , ax = plt.subplots(1,3,figsize=(20,8))
sns.barplot(x='Region', y= 'Sales', data=df , ax=ax[0]) 
sns.barplot(x='Region' , y= 'Profit', data=df , ax=ax[1])
sns.barplot(x='Region'  , y= 'Discount', data=df , ax=ax[2])
```

- Sales are higher in the South Region
- The Profit is higher in the West Region this may because of the hight Sales and low Discounts
- Central region has higher Discounts

                                        Based on Category

```python
fig , ax = plt.subplots(1,3,figsize=(20,8))
sns.barplot(x='Category' , y= 'Sales', data=df , ax=ax[0])
sns.barplot(x='Category' , y= 'Profit', data=df , ax=ax[1])
sns.barplot(x='Category' , y= 'Discount', data=df , ax=ax[2])
```

we can see clearly that even with the low Discounts the Technology Category has the higher Sales and Profits

                                      Based on Quantity

```python
fig , ax = plt.subplots(1,3,figsize=(24,5))
sns.barplot(x='Quantity' , y= 'Sales', data=df , ax=ax[0] )
sns.barplot(x='Quantity' , y= 'Profit', data=df , ax=ax[1])
sns.barplot(x='Quantity' , y= 'Discount', data=df , ax=ax[2])
```

                                       Based on Discount  

```python
fig,ax=plt.subplots(1,2,figsize=(20,5))
sns.lineplot(x='Discount', y='Sales', data = df,ax=ax[0])
sns.lineplot(x='Discount', y='Profit', data = df)
```

As the Discount is increasing the profits are going into negative. But the same trend can be observed with sales after 60% discount rates.At 50% discount the sales are maximum.

### Product Category Level Analysis

```python
#the percentage of every category in the dataset
plt.figure(figsize = (6,6))
textprops = {"fontsize":10}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index, autopct='%1.1f%%',textprops=textprops)
plt.show()
```

                               Quantity in each Sub-Category

```python
plt.figure(figsize=(8,8))
plt.title('Sub-Category')
plt.pie(df['Sub-Category'].value_counts(),labels=df['Sub-Category'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

As from the above Visualization,the most sold Sub-category items are Binders and Paper.

### Geographic Level Analysis

                                  State vs Sales

```python
plt.figure(figsize=(20,8))
df_state= df.groupby(['State'])[['Sales']].sum().sort_values('Sales' , ascending=False)
df_state.reset_index(inplace = True)
plt.xticks(rotation=90)
sns.barplot(x='State' , y='Sales' , data=df_state)
```

- California and Newyork has highest sales followed by Texas
- Maine , West Virgina and North Dakota has lowest sales

                                       State vs Profit

```python
plt.figure(figsize=(20,7))
df_state = df.groupby(['State'])[['Profit']].sum().sort_values('Profit' , ascending=False)
df_state.reset_index(inplace = True)
plt.xticks(rotation =90)
sns.barplot(x='State' , y='Profit', data=df_state )
```

- California has highest profits as well as sales followed by New York.

                                   State vs Discount

```python
plt.figure(figsize=(18,5))
data_state= df.groupby(['State'])[['Discount']].sum().sort_values("Discount", ascending=False) 
data_state.reset_index(inplace=True)
plt.xticks(rotation=90)
sns.barplot(x='State', y='Discount', data = data_state)
plt.show()
```

Maximum discount is provided in Texas where profits are in negative.

                         Analaysing Region with Profit and Sales

```python
pl=df.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pl[:].plot.bar(color=['blue','c'],figsize=(20,8))
plt.title('Profit/Loss and Sales across the Region')
plt.xlabel('Region')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

- The Highest Sales ans Profit gained in west region,followed by East region.
- South region takes lowest in sales,followed by central region which is having the least profit compare to other regions.

                          Analaysing States with Profit and Sales

```python
plt.figure(figsize=(20,8))
data_state= df.groupby(['State'])[['Sales', 'Profit']].sum().sort_values('Profit' , ascending=False)
data_state.reset_index(inplace = True)
data_state = data_state.melt('State')
plt.xticks(rotation=90)
sns.barplot(x='State', y='value', hue='variable', data=data_state)
```

California is the top city in terms of sales
